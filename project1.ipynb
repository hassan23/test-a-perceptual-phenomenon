#### Q1. What is our independent variable? What is our dependent variable?


**Answer**:
    
   Our **independent variable** is the **condition** of congruent words and incongruent words.
   
   Our **dependent variable** is the time it takes to read the color name. we can also call it the **performance** of each participant.
   <hr>

#### Q2. What is an appropriate set of hypotheses for this task? What kind of statistical test do you expect to perform? Justify your choices.

**Answer**:
    
   If **$μ_C$** is the mean of the population response with the congruent condition.
   
   and if **$μ_I$** is the mean of the population response with the incongruent condition.

   then,
   
   **the null hypothesis, $H_O$** : there is **no** significant difference between the response of population with congruent condition to the incongruent condition.
   <center>$μ_C - μ_I = 0$</center>
   
   **the alternate hypothesis, $H_A$**: there is a significant difference between the response of population with congruent condition to the incongruent condition.
   <center>$μ_C - μ_I \neq 0$</center>
   
   We will use **paired T-test** on the data, because
   <ul><li>
   We don't know the standard deviation of the population.
   </li>
   <li>
   The sample is paired, or two conditions are assigned to the same sample. 
   </li></ul>

#### Q5. Now, perform the statistical test and report your results. What is your confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?

**Answer:**

##### We are performed two-tailed dependent t-test for paired samples with 95% confidence Interval, Below are our results

**Degrees Of Freedom(df):** $23$<br>
**T-critical values at Alpha level($\alpha$) of 0.05 range: ** from $-2.069$ to $+2.069$<br>
**$P value<0.0001$**<br>
**Standers Error$(SE)$ of difference:** $0.993$<br>
**Margin Of Error $(ME)$ :** $2.054$<br>
**95% Confidence Interval($CI$):** from $-10.019$ to $-5.910$<br>
**T-statistic $T_S$:**$8.02$<br>

   
Here we see our T-statistic value is much **greater** than our **T-critical** value of $2.069$, so we are **rejecting** the null hypothesis **$(H_O)$**, which means that the incongruent condition does has an effect on response time of the participants in the sample.   
Yes I think the result is **statistically significant** and does meet up our expectation that there is **significant** difference in the response time of congruent and incongruent condition.
